```python
import torch
import ollama as lama

```

```python
prompt = "Is the sky pink?"
```

```python
stream = lama.chat(
model='gemma3:1b', # llama3.1:latest
messages=[{'role': 'user', 'content': prompt}],
stream=True,
)


for chunk in stream:
    print(chunk['message']['content'], end='', flush=True)
```
