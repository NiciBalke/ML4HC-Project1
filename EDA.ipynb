# Exploratory data analysis

## Loading Data and importing Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
import pandas as pd
import math
processed_df = pd.read_parquet("processedDataProxy.parquet")
print(processed_df.columns)
```

```python
def plot_vars(dataframe):
       vars_to_plot = ['Age', 'BUN', 'Creatinine', 'GCS', 'Gender', 'Glucose', 'HCO3',
              'HCT', 'HR', 'Height', 'ICUType', 'K', 'Mg', 'NIDiasABP', 'NIMAP',
              'NISysABP', 'Na', 'Platelets', 'RespRate', 'Temp',
              'TroponinT', 'Urine', 'WBC', 'Weight', 'ALP', 'ALT', 'AST',
              'Albumin', 'Bilirubin', 'FiO2', 'MechVent', 'PaCO2', 'PaO2', 'pH',
              'DiasABP', 'MAP', 'SaO2', 'SysABP', 'Lactate', 'Cholesterol',
              'TroponinI']
       cols = 4
       rows = math.ceil(len(vars_to_plot) / 4)
       fig, axes = plt.subplots(rows, cols,figsize= (cols*5, rows *4))
       axes = axes.flatten()

       for i, name in enumerate(vars_to_plot):
              ax = axes[i]
              data_to_plot = dataframe[name].dropna()
              # ax.set_title(f"distribution of {name}")
              sb.histplot(data_to_plot, kde = True, ax = ax)
       

```
